# **Library**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

# **A. Memanggil data**

```python
df = pd.read_excel("CAD-Dataset.xlsx")
```

## **Penjelasan :**

Membaca dataset dari file Excel menggunakan pandas. Dan variabelkan data menjadi df.

# **B. Menampilkan data**

```python
df.head(10)
```

```python
df.info()
```

## **Penjelasan :**

Melihat 10 baris pertama dengan `df.head(10)`, untuk memahami isi dataset. Kemudian, `df.info()` digunakan untuk melihat jumlah data, tipe data di setiap kolom, serta apakah ada data yang kosong.

# **C. Preprocessing data**

```python
  df.isnull().sum()
```

```python
# Menghapus kolom 'BMI' karena hampir semua datanya kosong
df = df.drop(columns=['BMI'])

# Mengisi nilai kosong pada kolom numerik ('FBS', 'ESR') dengan median
df['FBS'].fillna(df['FBS'].median(), inplace=True)
df['ESR'].fillna(df['ESR'].median(), inplace=True)

# Mengisi nilai kosong pada kolom kategori dengan kategori 'Unknown'
kategori_kosong = ['DM', 'Current Smoker', 'Nonanginal', 'LowTH Ang']
for col in kategori_kosong:
    df[col].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

## **Penjelasan :**

Kolom BMI dihapus karena sebagian besar datanya kosong, sementara kolom numerik seperti FBS dan ESR diisi dengan nilai median supaya tidak mengganggu distribusi data. Untuk kolom kategori seperti DM, Current Smoker, Nonanginal, dan LowTH Ang, nilai kosong diisi dengan kategori Unknown.

# **D. Mengubah semua kolom menjadi bernilai numerik**

```python
df.head()
```

```python
# Mengubah semua kolom kategori menjadi nilai numerik
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str)
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df.head()
```

## **Penjelasan :**

Mengubah semua kolom kategori menjadi numerik menggunakan LabelEncoder. Setiap kategori unik dalam suatu kolom akan dikonversi menjadi angka yang mewakili kategori tersebut supaya lebih mudah diolah.

# **E. Melakukan proses normalisasi pada masing-masing atribut**

```python
df.head(10)
```

```python
# Normalisasi semua fitur numerik agar berada dalam rentang 0-1
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])
```

```python
df.head(10)
```

## **Penjelasan :**

Normalisasi pada seluruh fitur numerik supaya nilainya berada dalam rentang 0-1 menggunakan MinMaxScaler. Normalisasi membuat lebih optimal karena menghindari skala yang terlalu besar atau kecil pada variabel tertentu.

# **Kesimpulan**

Setelah dilakukan preprocessing, dataset CAD telah bersih.
